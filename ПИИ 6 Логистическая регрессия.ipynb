```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ITMO_PRO_ML_Course_1/6/candy-data.csv', sep=',')
df = df.set_index('competitorname')
df = df.drop(['Candy Corn', 'Nerds', 'Reeses stuffed with pieces'], axis = 0)
df.head()
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ITMO_PRO_ML_Course_1/6/candy-test.csv', sep=',')
df1 = df1.set_index('competitorname')
df1
```

```python
y_train = df['Y']
X_train = df.drop(columns=['winpercent','Y'])

y_test = df1['Y']
X_test = df1.drop(columns=['Y'])

topredict1 = df1.loc[['Tootsie Roll Juniors'],['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']]
topredict2 = df1.loc[['Swedish Fish'],['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']]
```

```python
lr = LogisticRegression(random_state = 2019, solver = 'lbfgs').fit(X_train, y_train)
```

```python
print(lr.predict(X_test))
```

```python
print(lr.predict_proba(topredict1))
print(lr.predict_proba(topredict2))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lr.predict(X_test))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, lr.predict(X_test))
```

```python
from sklearn.metrics import recall_score
recall_score(y_test, lr.predict(X_test))
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, lr.predict(X_test))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, lr.predict_proba(X_test)[:, 1])
```
