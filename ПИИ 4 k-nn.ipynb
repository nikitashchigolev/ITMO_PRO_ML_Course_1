```python
import numpy as np
from scipy.spatial import distance
```

```python
import pandas as pd
import io

data = """id,X,Y,Class
1,50,60,0
2,93,51,0
3,55,62,1
4,26,56,0
5,66,81,0
6,18,37,0
7,18,82,0
8,72,13,0
9,98,96,0
10,68,65,1"""
df = pd.read_csv(io.StringIO(data), sep=",", index_col='id')
df
```

```python
x = np.array([[56,67]])
df['Euclidean'] = [distance.euclidean(obj, x) for index, obj in df.iloc[:,:2].iterrows()]
df['Manhattan'] = [distance.cityblock(obj, x) for index, obj in df.iloc[:,:2].iterrows()]
df
```

```python
print(min(df['Euclidean']))
print(min(df['Manhattan']))
```
