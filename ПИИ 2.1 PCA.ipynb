```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/14_16_2.csv', sep=';')
print(df)
```

```python
#МГК или PCA
pca = PCA(n_components=2, svd_solver='full')
principal_components = pca.fit_transform(df)

#доля объясненной дисперсии первых двух главных компонент
print(sum(pca.explained_variance_ratio_))

x_pca = pd.DataFrame(data = principal_components, columns = ['PC1','PC2'])
x_pca
```

```python
#группы объектов первых двух компонент
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
plt.scatter(x_pca.iloc[:,0], x_pca.iloc[:,1])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```
